```python
imagens = []
```

```python
#Carregar as 1776 imagens
z = 0
for i in range(2):
    for j in range(10):
        for l in range (10):
            
            for m in range(2):
                for n in range(10):
                    
                    a = int(f'{m}{n}')
                    b = int(f'{i}{j}{l}')

                    if a > 16 or b > 111 or (n == 0 and m == 0) or (i== 0 and j == 0 and l == 0):
                        pass
                    else:
                        #print('c{}{}{}_0{}{}.png'.format(i,j,l, m,n))
                        image_aux = cv2.imread("Brodatz128x128(Tex)/c{}{}{}_0{}{}.png".format(i,j,l,m,n))
                        imagens.append(image_aux[:128,:128,0])
                        z +=1
                    
print(z)
                    
```

```python
#print(a.ravel())transforma em um vetor a matriz
soma = []
media = []
minimo = []
traco = []
maximo = []
desvio = []
pico_a_pico = []

for i in range(len(imagens)):
    a = imagens[i]
    soma.append(a.sum())
    media.append(a.mean())
    minimo.append(a.min())
    traco.append(a.trace())
    maximo.append(a.max())
    desvio.append(a.std())
    pico_a_pico.append(a.ptp())

```

```python
k = []
for i in range(1,112):
    for j in range(1,17):
        k.append(i)
```

```python
colunas = [soma,media,minimo,traco,maximo,desvio,pico_a_pico,k]
save_arq = pd.DataFrame(colunas)
save_arq = save_arq.transpose()
save_arq
```

```python
save_arq.to_csv("matrix.csv")
```
